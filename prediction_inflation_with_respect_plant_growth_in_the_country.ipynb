```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"/content/test.csv")
df.head(10)
df.shape
df.describe()
df = df.dropna(axis = 0)
df.info()
```

```python
from datetime import datetime
x = df['date']
df['date'] = df['date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
df.info()
```

```python
df = df.drop(['ISO3'], axis=1)
df
```

```python
import pandas as pd
encoded_df = pd.get_dummies(data = df, columns =['country'])
encoded_df
```

```python
import pandas as pd
x_train = encoded_df.drop(labels = 'Inflation',axis = 1 )[encoded_df['date'].dt.year < 2020]
x_test = encoded_df.drop(labels = 'Inflation',axis = 1 )[encoded_df['date'].dt.year >= 2020]

y_train = encoded_df[encoded_df['date'].dt.year < 2020]['Inflation'].values
y_test = encoded_df[encoded_df['date'].dt.year >= 2020]['Inflation'].values

x_train.drop(labels = 'date',axis = True , inplace = True)
x_test.drop(labels = 'date',axis = True , inplace = True)

print(x_test)
print(x_train)
np.shape(y_train)
encoded_df
```

```python
!pip  install scikit-learn
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()
linear_regressor.fit(x_train,y_train)
y_pred_lr = linear_regressor.predict(x_test)
# y_pred_lr
from sklearn.metrics import mean_absolute_error as mae , mean_squared_error as mse
print("----------Linear Reggression- Model Evaluvation-----------")
print("Mean Absolute Error (MAE) : {}".format(mae(y_test,y_pred_lr)))
print("Mean Squared Error (MSE) : {}".format(mse(y_test,y_pred_lr)))
print("Root MeanSquared Error (RMSE) : {}".format(np.sqrt(mse(y_test,y_pred_lr))))
print(linear_regressor.score(x_test,y_test))
print(linear_regressor.score(x_train,y_train))
```

```python
from sklearn.tree import DecisionTreeRegressor
Decision_Regressor = DecisionTreeRegressor()
Decision_Regressor.fit(x_train,y_train)
y_pred_dr = Decision_Regressor.predict(x_test)
# y_pred_dr
print("----------Decision Tree Regressor- Model Evaluvation-----------")
print("Mean Absolute Error (MAE) : {}".format(mae(y_test,y_pred_dr)))
print("Mean Squared Error (MSE) : {}".format(mse(y_test,y_pred_dr)))
print("Root MeanSquared Error (RMSE) : {}".format(np.sqrt(mse(y_test,y_pred_dr))))
print(Decision_Regressor.score(x_test,y_test))
print(Decision_Regressor.score(x_train,y_train))
```

```python
from sklearn.ensemble import RandomForestRegressor
Random_Regressor = RandomForestRegressor()
Random_Regressor.fit(x_train,y_train)
y_pred_rr = Decision_Regressor.predict(x_test)
y_pred_rr
print("----------Random Forest Regressor - Model Evaluvation-----------")
print("Mean Absolute Error (MAE) : {}".format(mae(y_test,y_pred_rr)))
print("Mean Squared Error (MSE) : {}".format(mse(y_test,y_pred_rr)))
print("Root MeanSquared Error (RMSE) : {}".format(np.sqrt(mse(y_test,y_pred_rr))))
print(Random_Regressor.score(x_test,y_test))
print(Random_Regressor.score(x_train,y_train))
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse
import numpy as np
svm_regressor = SVR()
svm_regressor.fit(x_train, y_train)
y_pred_svm = svm_regressor.predict(x_test)
print("----------Support Vector Machine Regressor - Model Evaluation-----------")
print("Mean Absolute Error (MAE) : {}".format(mae(y_test, y_pred_svm)))
print("Mean Squared Error (MSE) : {}".format(mse(y_test, y_pred_svm)))
print("Root Mean Squared Error (RMSE) : {}".format(np.sqrt(mse(y_test, y_pred_svm))))
print("R² Score : {}".format(svm_regressor.score(x_test, y_test)))
print("Training R² Score : {}".format(svm_regressor.score(x_train, y_train)))
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse
import numpy as np
model = Sequential()
model.add(Dense(units=64, activation='relu', input_dim=x_train.shape[1]))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=1, activation='linear'))
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x_train, y_train, epochs=50, batch_size=32, validation_data=(x_test, y_test))
y_pred_nn = model.predict(x_test).flatten()
print("----------Neural Network Regressor - Model Evaluation-----------")
print("Mean Absolute Error (MAE) : {}".format(mae(y_test, y_pred_nn)))
print("Mean Squared Error (MSE) : {}".format(mse(y_test, y_pred_nn)))
print("Root Mean Squared Error (RMSE) : {}".format(np.sqrt(mse(y_test, y_pred_nn))))
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse
import numpy as np
knn_regressor = KNeighborsRegressor(n_neighbors=20)
knn_regressor.fit(x_train, y_train)
y_pred_knn = knn_regressor.predict(x_test)
print("----------K-Nearest Neighbors Regressor - Model Evaluation-----------")
print("Mean Absolute Error (MAE): {}".format(mae(y_test, y_pred_knn)))
print("Mean Squared Error (MSE): {}".format(mse(y_test, y_pred_knn)))
print("Root Mean Squared Error (RMSE): {}".format(np.sqrt(mse(y_test, y_pred_knn))))
print("R² Score : {}".format(knn_regressor.score(x_test, y_test)))
print("Training R² Score : {}".format(knn_regressor.score(x_train, y_train)))
```
